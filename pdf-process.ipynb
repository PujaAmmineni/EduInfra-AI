Remove PDF first & last page

```python
import os
import fitz  # PyMuPDF

# Folder containing PDFs
folder_path = "/Users/aakashwalavalkar/Downloads/CCE 4360"
output_folder = "/Users/aakashwalavalkar/Downloads/CCE 4360 processed"

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Process each PDF in the folder
for filename in os.listdir(folder_path):
    if filename.lower().endswith(".pdf"):
        input_pdf_path = os.path.join(folder_path, filename)
        output_pdf_path = os.path.join(output_folder, filename)

        # Open the PDF
        doc = fitz.open(input_pdf_path)
        total_pages = len(doc)

        # Create a new PDF excluding first and last page
        new_doc = fitz.open()
        for i in range(1, total_pages - 1):  # Exclude first (0) and last (total_pages-1)
            new_doc.insert_pdf(doc, from_page=i, to_page=i)

        # Save modified PDF
        if new_doc.page_count > 0:  # Ensure we don't save an empty PDF
            new_doc.save(output_pdf_path)
            print(f"Processed: {filename}")
        else:
            print(f"Skipped (too few pages): {filename}")

        # Close the documents
        doc.close()
        new_doc.close()

print("All PDFs processed successfully.")
```
